# MODEL IMPLEMENTATION:

```python
import os
import torch
import torch.nn as nn
import numpy as np

from torchvision.models import resnet18, ResNet18_Weights
```

```python
class Model(nn.Module):

  def __init__(self):
        super(Model, self).__init__()
        pretrained_model = resnet18(pretrained=True)
        self.backbone = nn.Sequential(*list(pretrained_model.children())[:-2])
        self.have_object = nn.Linear(25088, 1)
        self.bbox = nn.Linear(25088, 4)
        self.cat_or_dog = nn.Linear(25088, 1)
        self.specie = nn.Linear(25088, 9)

  def forward(self, input):
        out_backbone = self.backbone(input)
        out_backbone = out_backbone.flatten(start_dim=1)
        have_object = torch.relu(self.have_object(out_backbone))
        bbox = self.bbox(out_backbone)
        cat_or_dog = torch.relu(self.cat_or_dog(out_backbone))
        specie = torch.sigmoid(self.specie(out_backbone))

        return {
            "bbox": bbox,
            "object": have_object,
            "cat_or_dog": cat_or_dog,
            "specie": specie
        }
```

```python
from bs4 import BeautifulSoup

def read_xml_file(path):
    try:
        with open(path, 'r') as f:
            data = f.read()
        bs_data = BeautifulSoup(data, 'xml')

        name_elem = bs_data.find("name")
        xmin_elem = bs_data.find("xmin")
        ymin_elem = bs_data.find("ymin")
        xmax_elem = bs_data.find("xmax")
        ymax_elem = bs_data.find("ymax")

        if name_elem is None or xmin_elem is None or ymin_elem is None or xmax_elem is None or ymax_elem is None:
            return None

        return {
            "cat_or_dog": name_elem.text,
            "xmin": int(xmin_elem.text),
            "ymin": int(ymin_elem.text),
            "xmax": int(xmax_elem.text),
            "ymax": int(ymax_elem.text),
            "specie": "_".join(path.split(os.sep)[-1].split("_")[:-1])
        }
    except FileNotFoundError:
        return None
```

```python
from PIL import Image

import torchvision.transforms as transforms


class CustomDataset_old():

  def __init__(self, dataset_path, images_list, transform=None):
        self.dataset_path = dataset_path
        self.images_list = images_list
        self.transform = transform

  def __len__(self):
        return len(self.images_list)

  def __getitem__(self, index):
        image_name = self.images_list[index]
        image_path = os.path.join(self.dataset_path, "images", image_name + ".jpg")
        label_path = os.path.join(self.dataset_path, "labels", image_name + ".xml")

        image = Image.open(image_path).convert("RGB")
        label = read_xml_file(label_path)

        if self.transform is not None:
            image = self.transform(image)

        return image, label
```

```python
class CustomDataset(torch.utils.data.Dataset):

    def __init__(self, dataset_path, images_list, train=False):
        self.dataset_path = dataset_path
        self.images_list = images_list
        self.train = train

        self.preprocess = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
        ])

        self.image_folder_path = os.path.join(dataset_path, "images")
        self.label_folder_path = os.path.join(dataset_path, "labels")

        self.images = []
        self.labels = []

        for path in os.listdir(self.image_folder_path):
            name = path.split(os.sep)[-1].split(".")[0]
            if name in self.images_list:
                xml_path = os.path.join(self.label_folder_path, name+".xml")
                xml_data = read_xml_file(xml_path)
                self.images.append(path)
                self.labels.append(xml_data)

    def __len__(self):
        return len(self.images)

    def __getitem__(self, index):
      
        image_path = self.images[index]
        xml_path = os.path.join(self.label_folder_path, os.path.splitext(image_path)[0] + '.xml')
        xml_data = read_xml_file(xml_path)

        if xml_data is None:
            # Handle the case when XML data is not available
            return None

        image = Image.open(os.path.join(self.image_folder_path, image_path))
        image = self.preprocess(image)
        # Add an extra dimension to the image tensor
        image = image.unsqueeze(0)

        labels = {
            "cat_or_dog": xml_data["cat_or_dog"],
            "xmin": xml_data["xmin"],
            "ymin": xml_data["ymin"],
            "xmax": xml_data["xmax"],
            "ymax": xml_data["ymax"],
            "specie": xml_data["specie"]
        }

        return image, labels
```
